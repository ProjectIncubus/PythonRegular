```python
import pandas as pd # Importamos pandas como pd
import numpy as np # Importamos numpy como np
```

```python
# Asi podemos realizar la lectura de nuestros archivos.
pokedf_xlsx = pd.read_excel('pokemon_data.xlsx') # Si el archivo no está en el mismo folder que
# el script, tenemos que indicar el path completo.
pokedf_txt = pd.read_csv('pokemon_data.txt', delimiter='\t') # La lectura de archivos .txt se hace
# como .csv pero indicamos que el delimitador es el tabulador.

pokedf = pd.read_csv('pokemon_data.csv', index_col=0) # Trabajaremos con este archivo .csv
# index_col=0 sirve para indicarle a Pandas que el archivo csv ya cuenta con una columna
# de índices, y es la columna 0. Si no tuviera este argumento, Pandas crearía una columna
# antes de esta con los índices.

print(type(pokedf))
```

```python
# El método .head() nos muestra por default las primeras 5 filas de nuestro data frame.
# Si ponemos un argumento, será el número de filas a mostrar, en este caso 3.
pokedf.head(3)
```

```python
# Veremos que Pandas nos da la oportunidad de acceder a las columnas y a los valores de las
# columnas como si fueran los atributos del data frame.
print(pokedf.columns)
print(pokedf.columns.values)
print(type(pokedf.columns.values)) # Aquí vemos que la "lista" con los nombres de las columnas
# es más bien un array de numpy (que funciona muy parecido a una lista, con las ventajas de que
# se puede trabajar con él como un vector).
```

```python
columnas = pokedf.columns.values # Aquí creamos la variable columnas con este array de nombres de
# las columnas.

print(pokedf[columnas[0:4]]) # Aquí imprimimos el data frame indicando que solo queremos
# de la columna 0 a la 3. Recordar que escribimos 4 porque Pandas trabaja con el 4 exclusivo.

# ¿Te parece familiar? ¡Es porque usa el mismo modo de trabajo que un slicing o rebanado de
# listas o strings! :D
```

```python
print(pokedf[columnas]) # Si lo dejáramos así, indicamos que queremos todas las columnas.
```

```python
# También podemos acceder a una columna en específico por medio del nombre de dicha columna.
# En este caso columnas.
print(pokedf[['Name', 'Attack', 'Defense']])

poke_hp = pokedf[['HP']] # Aquí creamos una variable que almacena solo la columna "HP"
print(type(poke_hp)) # Si consultamos el tipo veremos que es aún un data frame de Pandas.

poke_hp.head()
```

### Acceso a valores del DF por  medio de indices

```python
pokedf.head(3)
```

```python
# Para acceder a un valor específico del data frame utilizamos
# el atributo iloc e indicamos los valores en formato [fila, columna]
print(pokedf.iloc[0, 0])

# También podemos acceder a un rango de filas y columnas con un
# rebanado.
print(pokedf.iloc[0:4,0:4])
print(pokedf.iloc[:,0:4]) # Todas las filas, de la columna 0 a 3.
print(pokedf.iloc[0:4,:]) # Las filas 0 a 3, todas las columnas.
```

### Cambios en el DF

```python
pokedf.head(3)
```

```python
# Aquí utilizamos el método .copy() para copiar pokedf en un data
# frame nuevo llamado pokedf_copy. Esto es muy útil cuando sabemos
# que vamos a estar modificando el data frame y no queremos que se
# estropee el original.
pokedf_copy = pokedf.copy()

# Ahora indicamos que si no existe la columna "Total Stats" CREALA
# y que contenga una suma horizontal para todas las filas de la
# columna 3 a la 8 (recordemos que el 9 se excluye).
pokedf_copy['Total Stats'] = pokedf_copy.iloc[:,3:9].sum(axis=1)

# Podemos entender la línea de arriba como:
#   Con .iloc hazme un subset del data frame pokedf_copy que incluya
#      todas las filas y las columnas 3 a 8.
#   Luego, con el método .sum() indico que quiero que se haga una suma
#      de ese subset, pero como argumento indico que quiero que la suma
#      sea en el eje 1, es decir horizontal. (Si quisiéramos que la suma
#      fuera vertical, el argumento sería axis=0).
#      Queremos que la suma sea horizontal porque queremos sumar POR FILA
#      el valor en la columna HP, Attack, Defense, etc.
#   Y ya esa suma la asignamos a una columna llamada "Total Stats",
#       si la columna no existe, se crea.
```

```python
pokedf_copy.head() # Aqui vemos que la columna nueva se va al final del DF.
```

```python
# ¿Qué haríamos si quisieramos mover la columna a la izquierda de la columna
# "HP"?

# Creamos una variable cols que almacena una lista con los nombres
# de las columnas.
cols = list(pokedf_copy.columns.values)
cols_error = pokedf_copy.columns.values
# ¿Por qué convertimos el array de numpy a una lista?
# Para poder realizar la concatenación de abajo.
print(type(cols[0:3])) # Tipo lista
print(type(cols[-1])) # Tipo string, porque es solo un elemento de la lista
# cols
print(type(cols[3:-1])) # Tipo lista

# Realizamos el reordenamiento:
# El nuevo pokedf_copy es el pokedf_copy PERO con la lista cols REBANADA
#   primero una rebanada con los elementos 0 a 2, luego el último que es
#   la columna "Total Stats", y luego el resto de columnas.
pokedf_copy = pokedf_copy[cols[0:3] + [cols[-1]] + cols[3:11]]
# ¿Por qué cols[-1] lo metimos en otros corchetes y los rebanados no?
# Porque, si vemos en la parte de arriba, donde consultamos el tipo,
# cols[-1] NO ES UNA LISTA, sino un string, y al meterlo en [] ahora es
# una lista. Si no lo hacemos no nos deja concatenar, porque no podemos
# concatenar una lista + string, solo lista + lista.

pokedf_copy.head()
```

```python
# Si no realizamos la conversion a lista, nos marca un error,
# ya que cols seria un array de numpy.

print(type(cols_error[0:3])) # Array de numpy
print(type(cols_error[-1])) # ESTO SIGUE SIENDO UN STRING, YA QUE ES
# UN ELEMENTO DE ARRAY DE NUMPY.
print(type(cols_error[3:-1])) # Array de numpy

print(cols_error[0:3])
print(cols_error[-1])
print(cols_error[3:-1])

# Si corremos la linea de abajo descomentada, nos marca el siguiente error:
#   ValueError: operands could not be broadcast together with shapes (3,) (8,)
#   ya que la concatenación de arrays NO funciona como la concatenación
#   de listas. Es por eso que debemos convertir el array a una lista.
# print(cols_error[0:3] + cols_error[-1] + cols_error[3:-1])
```

### Aggregate Statistics (groupby)

```python
# El método .groupby() nos sirve para realizar agrupaciones de acuerdo
# a un argumento. En este caso agrupamos por Type 1 y Type 2 las columnas
# 'Total Stats', 'HP', 'Attack', 'Defense'. Ya realizado el agrupamiento
# debemos indicar qué queremos hacer, en este caso con el método .mean()
# indicamos que a la agrupación le queremos sacar su promedio.

pokedf_copy.groupby(['Type 1', 'Type 2'])[['Total Stats', 'HP', 'Attack', 'Defense']].mean()
```

```python
# Aquí agrupamos solo por Type 1.
pokedf_copy.groupby('Type 1')[['Total Stats', 'HP', 'Attack', 'Defense']].mean()
```

```python
# Aquí agrupamos y sacamos desviación estándar con el método .std()
pokedf_copy.groupby(['Type 1', 'Type 2'])[['Total Stats', 'HP', 'Attack', 'Defense']].std()
```
